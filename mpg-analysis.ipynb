```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
mpg = sns.load_dataset("mpg")
mpg.head(10)
```

### Create a visual showing the relationship between horsepower and fuel economy

```python
sns.relplot(data=mpg,
x='horsepower', y='mpg')
plt.xlabel("Horsepower")
plt.ylabel("mpg")
plt.title("Relationship between Horsepower and Fuel Economy")
```

##### <font color="red">As the power of the car (horsepower) increases, fuel economy generally appears to decrease. Therefore, there is a negative correlation between horsepower and fuel economy.</font>

### Create a table summarizing fuel economy by country of origin

```python
fuel_economy_by_origin = mpg.groupby(by="origin")["mpg"].mean().reset_index()
fuel_economy_by_origin = fuel_economy_by_origin.rename(columns={"mpg":"average mpg"})
fuel_economy_by_origin
```

### Create a visual showing how fuel economy has changed over time

```python
fuel_economy_over_time = mpg.groupby(by="model_year")["mpg"].mean().reset_index()
sns.lineplot(data=fuel_economy_over_time, x="model_year", y="mpg")
plt.title("Fuel Economy Over Time")
plt.xlabel("Year")
plt.ylabel("Average mpg")
```

##### <font color="red"> From 1975, there was an upward trend, and from 1980, there was a slight decline. This often has a direct correlation with historical events and measures</font>

### Create two other visuals or tables related to fuel economy

```python
# Create a barplot showing the average fuel economy (MPG) per number of cylinders
sns.barplot(data=mpg, x="cylinders", y="mpg", errorbar= None)
plt.title("Fuel Economy By Number Of Cylinders")
plt.xlabel("Number of Cylinders")
plt.ylabel("Average mpg")
```

##### <font color="red"> The highest average MPG was observed in cars with 4 cylinders. As the number of cylinders increases thereafter, the average MPG decreases. Therefore, the optimal fuel efficiency is at 4 cylinders </font>

```python
# Create a histogram of the car weights distributed across the different countries of origin
sns.histplot(data=mpg, x="weight", hue="origin")
plt.title("Distribution of Car Weight by Country of Origin")
plt.xlabel("Weight")
plt.ylabel("Frequency")
```

##### <font color="red"> In Japan, primarily cars with lower weight are distributed. This also applies to Europe, but in Europe there is also a small number of cars with higher weight (up to 4000). The USA is by far the country where the most cars are distributed with higher weight </font>
